### In pandas, the df.isnull() function is used to check for missing or null values in a DataFrame. It returns a DataFrame of the same shape as the input DataFrame, where each element is a boolean value indicating whether the corresponding element in the original DataFrame is null or not.

### syntax = df.isnull()

```python
df.isnull()
```

### In pandas, the df.fillna() function is used to fill missing or null values in a DataFrame with specified values. It allows you to replace NaN (null) values with a scalar value, a specified DataFrame, or with values derived from a function or another DataFrame.

### syntax = df.fillna()

```python
df.fillna(0)
```

### In pandas, the df.nlargest() function is used to retrieve the rows with the largest values in one or more columns of a DataFrame. It allows you to extract the top n rows based on a specified column(s) and sort order.

### In pandas, the df.copy() function is used to create a deep copy of a DataFrame. It returns a new DataFrame that is a separate and independent copy of the original DataFrame. Changes made to the copied DataFrame will not affect the original DataFrame, and vice versa.

### syntax = df.copy()

```python

df_copy=df.copy()
df_copy['Quantity']= df_copy['Quantity'] * 2

print(df)
print(df_copy)


```
